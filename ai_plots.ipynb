```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Plotting settings
plt.rcParams['figure.figsize'] = (8, 6)
plt.rcParams['figure.dpi'] = 100
plt.rcParams['font.size'] = 14
plt.rcParams['axes.labelsize'] = 13
plt.rcParams['axes.titlesize'] = 13
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['font.family'] = 'serif'
plt.style.use('dark_background')

sns.set_style("white")
```

```python
df = pd.read_csv("results.csv")
```

```python
melted = df[["a_star_tree_h2_time", "a_star_graph_h2_time", "a_star_tree_h3_time", "a_star_graph_h3_time"]].melt()
```

```python
sns.boxplot(x="variable", y="value",
            data=melted)
plt.ylabel("Time, sec")
plt.xlabel("")
plt.title("Runtime Distribution of A* Search")
plt.xticks([0, 1, 2, 3], ["A* Tree H2", "A* Graph H2", "A* Tree H3", "A* Graph H3"])
plt.show()
```

```python
ax = sns.boxplot(x="variable", y="value",
            data=melted)
plt.ylim([0, 0.5])
plt.ylabel("Time, sec")
plt.xlabel("")
plt.title("Runtime Distribution of A* Search")
plt.xticks([0, 1, 2, 3], ["A* Tree H2", "A* Graph H2", "A* Tree H3", "A* Graph H3"])
plt.show()
```
